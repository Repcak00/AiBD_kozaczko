Ćwiczenie zostało wykonane na pierwszych laboratoriach AiBD. Poniżej widoczne są 3 wykresy wykonane dla zadanych argumentów

```python
import numpy as np
import matplotlib.pyplot as plt

function = lambda x: x**2 + 5

x1 = np.linspace(-1, 1)
x2 = np.linspace(-6, 6)
x3 = np.linspace(0, 5)

y1 = function(x1)
y2 = function(x2)
y3 = function(x3)
```

## Przedział nr 1: 

$x>-1$ oraz $x<1$

```python
plt.plot(x1, y1)
plt.ylabel('wartosci')
plt.xlabel('argumenty')
plt.title("Funkcja x^2 + 5 dla 1. przedzialu")
plt.legend(['x^2+5'])
plt.show()
```

## Przedział nr 2:

$x>-6$ oraz $x<6$

```python
plt.plot(x2, y2)
plt.ylabel('wartosci')
plt.xlabel('argumenty')
plt.title("Funkcja x^2 + 5 dla 2. przedzialu")
plt.legend(['x^2+5'])
plt.show()
```

## Przedział nr 3:

$x>0$ oraz $x<5$

```python
plt.plot(x3, y3)
plt.ylabel('wartosci')
plt.xlabel('argumenty')
plt.title("Funkcja x^2 + 5 dla 3 przedzialu")
plt.legend(['x^2+5'])

plt.show()
```
